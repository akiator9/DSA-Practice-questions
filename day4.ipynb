```python
N = 5
mid = N // 2

for i in range(N):
    if i <= mid:
        stars = 2 * i + 1
    else:
        stars = 2 * (N - i - 1) + 1

    spaces = (N - stars) // 2
    print(' ' * spaces + '*' * stars)
```

```python
N = 4

for i in range(1, N + 1):
    print(' ' * (N - i), end='')
    for j in range(i, i + i):
        print(j, end='')

    for j in range(i + i - 2, i - 1, -1):
        print(j, end='')

    print()
```
