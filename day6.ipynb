```python
'''PRINT Pattern for N = 3
    1=1
    1+2 = 3
    1+2+3 = 6
 
DO it for N = 5 '''
```

```python
N = 5
def pattern(num):
    ls = []
    while num > 0:
        ls.append(num)
        num = num-1
    return(ls)
    
for i in range(1,N+1):
    val = pattern(i)
    result = "+".join(map(str, val))
    # print(result)
    sm = sum(val)
    print(f"{result}={sm}")

```
